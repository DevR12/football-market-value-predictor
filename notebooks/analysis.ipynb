**Data Cleaning**

*Data Type Validation*
The numerical columns identified above -> numeric values, no strings

---

*Range Validation*

Players stats need to be within realistic ranges for pro football
- age between 16 and 45
- overall_rating between 0 and 100
- potential_rating between 0 and 100
- goals >= 0
- assists >= 0
- matches_played >= 0
- minutes_played >= 0
- market_value_millino_eur >= 0

Violating rows checked or removed

---

*Categorical Columns*

May have inconsistent formatting or whitespace

- removed leading/trailing whitepace
- consistent capitalization
- validating position labels

---

*Logical Checks*

- Players with 0 matches played should have 0 goals and assists
- Players with very low minutes played shouldn' have unusually high goals or assists
- Minutes played sould correlate with the matches played

---

*Duplicate Records*

Each player should appear only once.
Checed using player_id column, duplicates removed.
